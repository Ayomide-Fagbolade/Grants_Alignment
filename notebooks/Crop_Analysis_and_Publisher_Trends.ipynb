```python
import ast

def parse_crops_discussed(text):
    if pd.isna(text):
        return []  # Return an empty list for NaN values
    try:
        return ast.literal_eval(text)
    except (ValueError, SyntaxError):
        return []  # Return empty list for malformed strings

combined_dataset['crops_discussed'] = combined_dataset['crops_discussed'].apply(parse_crops_discussed)

display(combined_dataset[['Link', 'Crops', 'crops_discussed']].head())
```

### Normalizing All Discussed Crops

To address the variations in crop names, we'll create a mapping from common alternative names and misspellings to our defined `target_crops`. This will help in getting a more consistent list of *all* unique crops mentioned, not just those that were previously identified as `target_crops`.

```python
import ast

# Define a more comprehensive mapping for target crop variations
crop_normalization_map = {
    'sweet potato': 'sweet potato',
    'sweet potatoes': 'sweet potato',
    'orange flesh sweet potato': 'sweet potato',
    'orange fleshed sweet potato (ofsp)': 'sweet potato',
    'orange sweet potato': 'sweet potato',
    'white-fleshed sweet potato': 'sweet potato',
    'banana': 'banana/plantain',
    'bananas': 'banana/plantain',
    'plantain': 'banana/plantain',
    'plantains': 'banana/plantain',
    'banana/plantain': 'banana/plantain', # Already normalized, but good to include
    'matoke': 'banana/plantain',
    'matoke (banana/plantain)': 'banana/plantain',
    'yam': 'yam',
    'yams': 'yam',
    'white yam': 'yam',
    'yellow yam': 'yam',
    'cassava': 'cassava',
    'manioc': 'cassava',
    'cassava (yam)': 'cassava',
    'sorghum': 'sorghum',
    'white sorghum': 'sorghum',
    'guinea corn': 'sorghum',

}

# Extend target_crops to include this new grouping if desired for analysis, or keep separate
# For this task, we will focus on normalizing what was *found* to the original target_crops or defined groups.

def apply_normalization(crop_list_of_dicts, normalization_map):
    normalized_list = []
    if not crop_list_of_dicts:
        return []
    for item in crop_list_of_dicts:
        if isinstance(item, dict) and 'crop' in item:
            original_crop = item['crop'].lower() # Normalize to lower case for mapping
            # Check if the original crop or its part exists in the normalization map
            normalized_crop = normalization_map.get(original_crop, original_crop) # Default to original if not found
            normalized_list.append(normalized_crop)
    return sorted(list(set(normalized_list)))

# Create a new column for all normalized discussed crops
combined_dataset['all_normalized_discussed_crops'] = combined_dataset['crops_discussed'].apply(lambda x: apply_normalization(x, crop_normalization_map))

combined_dataset['all_normalized_discussed_crops']
```

```python
def check_and_extract_normalized_target_crops(normalized_crop_list, target_crops_list):
    """
    Checks if any crop in normalized_crop_list is in target_crops_list and extracts them.
    Returns a tuple: (boolean indicating if any target crop is found, list of target crops found).
    Assumes normalized_crop_list contains strings (e.g., from 'all_normalized_discussed_crops').
    """
    found_crops = []
    if not normalized_crop_list:
        return False, []
    for crop_name in normalized_crop_list:
        if crop_name in target_crops_list:
            found_crops.append(crop_name)
    found_crops = sorted(list(set(found_crops)))
    return len(found_crops) > 0, found_crops

# Apply the new function to the 'all_normalized_discussed_crops' column
combined_dataset[['has_target_crop_discussed', 'discussed_target_crops']] = \
    combined_dataset['all_normalized_discussed_crops'].apply(
        lambda x: check_and_extract_normalized_target_crops(x, target_crops)
    ).apply(pd.Series)

# Display the relevant columns to verify the changes
display(combined_dataset[['Link', 'Crops', 'crops_discussed', 'all_normalized_discussed_crops', 'has_target_crop_discussed', 'discussed_target_crops']].head())
```

```python
def get_target_crop_discussions_with_details(crops_discussed_list, normalization_map, target_crops_list):
    """
    Filters and normalizes the list of crop dictionaries, keeping only those
    that are target crops, and updates the crop name to its normalized form.
    """
    if not crops_discussed_list:
        return []

    result_list = []
    for crop_dict in crops_discussed_list:
        if isinstance(crop_dict, dict) and 'crop' in crop_dict:
            original_crop_name = crop_dict['crop'].lower()
            normalized_crop_name = normalization_map.get(original_crop_name, original_crop_name)

            if normalized_crop_name in target_crops_list:
                # Create a copy to modify without affecting the original
                modified_crop_dict = crop_dict.copy()
                modified_crop_dict['crop'] = normalized_crop_name
                result_list.append(modified_crop_dict)
    return result_list

# Apply this new function to create 'target_crop_discussions'
combined_dataset['target_crop_discussions'] = combined_dataset['crops_discussed'].apply(
    lambda x: get_target_crop_discussions_with_details(x, crop_normalization_map, target_crops)
)

# Display some examples to verify the new column
display(combined_dataset[combined_dataset['has_target_crop_discussed']][['Link', 'all_normalized_discussed_crops', 'discussed_target_crops', 'target_crop_discussions']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure 'Date' column is in datetime format
combined_dataset['Year'] = pd.to_datetime(combined_dataset['Date'], errors='coerce').dt.year

# Drop rows where Year is NaN (if any date conversion failed)
combined_dataset.dropna(subset=['Year'], inplace=True)
combined_dataset['Year'] = combined_dataset['Year'].astype(int)

# Prepare data for time series plot using 'discussed_target_crops'
time_series_data = []
for index, row in combined_dataset.iterrows():
    year = row['Year']
    for crop in row['discussed_target_crops']: # Use the new discussed_target_crops
        time_series_data.append({'Year': year, 'Crop': crop})

# Create a DataFrame from the collected data
time_series_df = pd.DataFrame(time_series_data)

# Count occurrences of each crop per year
annual_crop_counts = time_series_df.groupby(['Year', 'Crop']).size().reset_index(name='Count')

# Plotting the time series
plt.figure(figsize=(15, 8))
sns.lineplot(data=annual_crop_counts, x='Year', y='Count', hue='Crop', marker='o')
plt.title('Time Series of Normalized Discussed Target Crops')
plt.xlabel('Year')
plt.ylabel('Number of Discussions')
plt.grid(True)
plt.legend(title='Normalized Crop')
plt.tight_layout()
plt.show()
```

### Individual Time Series Plots for Each Normalized Target Crop (using `discussed_target_crops`)

```python
unique_crops = annual_crop_counts['Crop'].unique()

for crop_name in unique_crops:
    plt.figure(figsize=(12, 6))
    # Filter data for the current crop
    crop_data = annual_crop_counts[annual_crop_counts['Crop'] == crop_name]
    sns.lineplot(data=crop_data, x='Year', y='Count', marker='o')
    plt.title(f'Time Series of Discussion Frequency for {crop_name}')
    plt.xlabel('Year')
    plt.ylabel('Number of Discussions')
    plt.grid(True)
    # Set x-ticks to display all years and rotate labels
    plt.xticks(annual_crop_counts['Year'].unique(), rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd

# Create a temporary DataFrame with 'Link' and 'target_crop_discussions'
# Using combined_dataset_with_target_crops as it already filters for relevant articles
df_temp = combined_dataset_with_target_crops[['Link', 'target_crop_discussions']].copy()

# Explode the 'target_crop_discussions' list to get one row per crop discussion
df_exploded_crops = df_temp.explode('target_crop_discussions')

# Filter out rows where 'target_crop_discussions' became None after explode (if any empty lists existed)
df_exploded_crops.dropna(subset=['target_crop_discussions'], inplace=True)

# Extract 'crop' and 'issues' from the dictionary
df_exploded_crops['crop_name'] = df_exploded_crops['target_crop_discussions'].apply(lambda x: x.get('crop'))
df_exploded_crops['issues'] = df_exploded_crops['target_crop_discussions'].apply(lambda x: x.get('issues', []))

# Explode the 'issues' list to get one row per issue associated with a crop discussion
df_exploded_issues = df_exploded_crops.explode('issues')

# Extract 'issue_type' and 'sentiment' from the issue dictionaries
# Handle cases where 'issues' might be None or not a dict after exploding an empty list
df_exploded_issues['issue_type'] = df_exploded_issues['issues'].apply(lambda x: x.get('issue') if isinstance(x, dict) else None)
df_exploded_issues['sentiment'] = df_exploded_issues['issues'].apply(lambda x: x.get('sentiment') if isinstance(x, dict) else None)

# Create the new 'issue_type_and_sentiment' column as a dictionary
df_exploded_issues['issue_type_and_sentiment'] = df_exploded_issues.apply(
    lambda row: {'issue_type': row['issue_type'], 'sentiment': row['sentiment']} if pd.notna(row['issue_type']) or pd.notna(row['sentiment']) else None,
    axis=1
)

# Select and reorder the desired columns for the final CSV
final_target_crop_discussions_df = df_exploded_issues[[
    'Link',
    'crop_name',
    'issue_type_and_sentiment'
]].copy()

# Fill any potential NaN values in the new combined column
final_target_crop_discussions_df.fillna({'issue_type_and_sentiment': 'N/A'}, inplace=True)

# Save the resulting DataFrame to a CSV file
output_filename = 'target_crop_discussions_flattened.csv'
final_target_crop_discussions_df.to_csv(output_filename, index=False)

print(f"Successfully saved flattened target crop discussions to '{output_filename}'")
display(final_target_crop_discussions_df.head())
```
