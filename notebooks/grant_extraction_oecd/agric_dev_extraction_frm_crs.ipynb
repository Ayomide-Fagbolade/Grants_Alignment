```python
# Use Polars to read the Parquet file lazily
df = pl.scan_parquet('C:/Users/ayo/Downloads/CRS-parquet (1)/CRS.parquet')
# Collect to get a DataFrame and print column names
schema = df.collect().columns
print(schema)
```

```python
agric_crs = df.filter(pl.col('sector_name').str.contains('Agriculture', literal=True))  
```

```python
keeplist=list(set(['year','donor_code','donor_name','agency_name', 'crs_id', 'project_number',
          'initial_report','recipient_code','de_recipientcode','recipient_name','region_code',
 'de_regioncode','initial_report','region_name', 'incomegroup_name', 'flow_code',
          'flow_name','bi_multi', 'category', 'finance_t', 'aid_t', 'usd_commitment', 'usd_disbursement',
           'usd_commitment_defl', 'usd_disbursement_defl','commitment_date','expected_start_date',
 'completion_date', 'long_description','keywords',  'channel_code','channel_name','channel_reported_name','parent_channel_code','geography','ld_cflag',
 'ld_cflag_name','sector_name','recipient_code',
 'de_recipientcode',
 'recipient_name',
 'region_code',
 'de_regioncode','short_description',
 'project_title',
 'purpose_code',
 'purpose_name','short_description',
 'project_title',
 'purpose_code',
 'long_description',]))
```

```python
agric_crs.columns
```

```python
agric_crs.describe()
```

```python
slim_agric_crs_df = agric_crs.select(keeplist)
```

```python
slim_agric_crs_df.select(['finance_t']).unique().collect()['finance_t'].to_list()
```

```python
agric_std_grants_df= slim_agric_crs_df.filter(pl.col('finance_t')==110)
```

```python
agric_std_grants_df.select(['finance_t']).collect().shape
```

```python
#110 is code for standard grants
agric_std_grants_df.select(['finance_t']).collect().unique()
```

```python
# 
agric_std_grants_df.select(['aid_t']).collect()['aid_t'].unique().to_list()
```

```python
# 30 is code for private developmet finance
agric_private_dev_df=agric_std_grants_df.filter(pl.col('flow_code')==30)
```

```python
agric_private_dev_df.select(pl.nth(0)).collect().shape
```

```python
agric_private_dev_df.select(['initial_report']).collect()['initial_report'].value_counts()
```

```python
# group by year, initial_report, project_code and count rows
grouped = agric_private_dev_df.group_by(['year', 'project_number','initial_report']).agg(pl.len()).collect().sort(by=['year','project_number'])
print(grouped)
```

```python
agric_private_dev_df_with_commit= agric_private_dev_df.filter(pl.col('usd_commitment')>0)
```

```python
agric_private_dev_df_with_commit.select(pl.nth(0)).collect().shape
```

```python
#we are  only concerned with rows where commitents are made or modified
agric_private_dev_df_with_commit.select(['initial_report']).collect()['initial_report'].value_counts()
```

```python
final_csv=agric_private_dev_df_with_commit.collect()
```

```python
# Save the filtered dataframe to a CSV file in the grants_data directory
import os
os.makedirs("../grants_data", exist_ok=True)
final_csv.write_csv("../grants_data/agric_private_dev_df_with_commit.csv")
print("Data written to ../grants_data/agric_private_dev_df_with_commit.csv")
```
