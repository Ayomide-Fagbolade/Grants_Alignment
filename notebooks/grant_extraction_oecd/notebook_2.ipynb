```python
# Expanded crop list with names, groups, and variations combined
augmented_crops = [
    {
        "name": "cassava",
        "group": "tuber",
        "lang_variations": [
            "cassava", "koko", "manioc", "tuber", "mandioca", "yuca", 
            "manihot", "manihot esculenta", "cassave", "tubercle", "tubérculo", 
            "knolle", "root crop"
        ]
    },
    {
        "name": "sorghum",
        "group": "grain",
        "lang_variations": [
            "sorghum", "sorgo", "grain", "cereal", "milo", "great millet", 
            "sorgho", "moench", "sorghum bicolor", "shalu", "céréale", 
            "getreide", "cereales"
        ]
    },
    {
        "name": "banana",
        "group": "fruit",
        "lang_variations": [
            "banana", "plantain", "musa", "fruit", "banane", "banano", 
            "plátano", "platano", "musa sapientum", "fruta", "frucht"
        ]
    },
    {
        "name": "plantain",
        "group": "fruit",
        "lang_variations": [
            "plantain", "banana", "musa", "fruit", "banane plantain", 
            "plátano macho", "platano macho", "musa paradisiaca", "fruta", "frucht"
        ]
    },
    {
        "name": "potato",
        "group": "root",
        "lang_variations": [
            "potato", "tuber", "root", "tubér", "patata", "pomme de terre", 
            "kartoffel", "solanum tuberosum", "papa", "racine", "raíz", "wurzel"
        ]
    },
    {
        "name": "yam",
        "group": "root",
        "lang_variations": [
            "yam", "dioscorea", "root", "igname", "ñame", "yamswurzel", 
            "dioscorea alata", "racine", "raíz", "wurzel"
        ]
    }
]
```

```python
# Function to find crops in a text
def find_crops(text, crops):
    found = []
    if pd.isna(text):
        return found
    text_lower = text.lower()
    for crop in crops:
        for var in crop["lang_variations"]:
            if var.lower() in text_lower:
                found.append(var)
                break
    return list(set(found))

# Apply to short_description and long_description
df['found_crops_short'] = df['short_description'].apply(lambda x: find_crops(x, augmented_crops))
df['found_crops_long'] = df['long_description'].apply(lambda x: find_crops(x, augmented_crops))

# Combine results
df['found_crops'] = df.apply(lambda row: list(set(row['found_crops_short'] + row['found_crops_long'])), axis=1)

# Summary
print("Total rows with any crop mention:", df['found_crops'].apply(lambda x: len(x) > 0).sum())
print("Example rows:")
print(df[df['found_crops'].apply(lambda x: len(x) > 0)][['project_number', 'found_crops']].head())
```

```python
# Create a DataFrame containing only rows that mention a crop
select_crop_df = df[df['found_crops'].apply(lambda x: len(x) > 0)].copy()
# Save to grants_data folder
import os
os.makedirs("../../grants_data", exist_ok=True)
select_crop_df.to_csv("../../grants_data/select_crop_df.csv", index=False)
print("Saved", len(select_crop_df), "rows to ../../grants_data/select_crop_df.csv")
```
